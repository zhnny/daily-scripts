```python
import requests
```

```python
header = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.114 Safari/537.36 Edg/91.0.864.59"}
```

```python
res = requests.get(url, headers=header, timeout=5)
```

```python
res.encoding = "GBK"
```

```python
res.text
```

```python
from lxml import etree
```

```python
selector = etree.HTML(res.text) #将源码转化为能被XPath匹配的格式
```

```python
data = selector.xpath("//div[@class='tpc_cont']")[0]
```

```python
mag = data.xpath(("./a/text()"))
```

```python
with open('m.txt', 'a+', encoding='utf-8') as file:
    for i in range(0, len(mag)):
        file.write(mag[i]+'\n')
```
