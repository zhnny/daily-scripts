```python
import csv
```

```python
data = []
with open('data.txt') as csv_file:
    csv_reader = csv.reader(csv_file, delimiter=',')
    line_count = 0
    for row in csv_reader:
        line_count += 1
        temp = []
        for i in range(len(row)-1):
            temp.append(float(row[i]))
        data.append(temp)
    print(f'Processed {line_count} lines.')
data
```

```python
num_count = len(data)
num_count
```

```python
def calc_l0(point, arr):
    distance = []
    sum = 0
    for i in range(len(arr)):
        dis = pow(pow(point[1]-arr[i][1],2)+ pow(point[2]-arr[i][2],2),0.5)
        if(dis == 0):
            continue
#         print(dis)
        if(len(distance)<5):
            distance.append(dis)
        else:
            for j in range(len(distance)):
                if(dis<distance[j]):
                    distance[j]=dis
                    break
    for i in range(len(distance)):
        sum = sum + distance[i]
    return sum/len(distance)
    
```

```python
def calc_dis(point, arr):
    distance = []
    for i in range(len(arr)):
        dis = pow(pow(point[1]-arr[i][1],2)+ pow(point[2]-arr[i][2],2),0.5)
        if(dis == 0):
            continue
        distance.append(dis)
    return distance
```
