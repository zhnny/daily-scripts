```python
table_data = data.xpath("./div")
for n in range(0,len(table_data)):
    day = table_data[n]
    print("===============================")
    day_table = day.xpath("./div/table")
    for i in range(1,len(day_table)):
        time = day_table[0].xpath("./tr")[0].xpath("./td")[2]
        city = day_table[i]
        country = city.xpath("./tr")
        for j in range(0,len(country)):
            country_info = country[j].xpath("./td")
            country_info_list = []
            if j==0:
                country_info_list.append(country_info[0].text)
                country_info_list.append(country_info[1].xpath("./a/text()")[0])
                country_info_list.append(country_info[2].text)
                country_info_list.append(country_info[3].xpath("./span[1]/text()")[0])
                country_info_list.append(country_info[3].xpath("./span[2]/text()")[0])  
                country_info_list.append(country_info[4].text)
                country_info_list.append(country_info[5].text)
                country_info_list.append(country_info[6].xpath("./span[1]/text()")[0])
                country_info_list.append(country_info[6].xpath("./span[2]/text()")[0])  
                country_info_list.append(country_info[7].text)
                country_info_list.append(time.text[3:-3])
                print(country_info_list)
            else:
                country_info_list.append(country[0].xpath("./td")[0].text)
                country_info_list.append(country_info[0].xpath("./a/text()")[0])
                country_info_list.append(country_info[1].text)
                country_info_list.append(country_info[2].xpath("./span[1]/text()")[0])
                country_info_list.append(country_info[2].xpath("./span[2]/text()")[0])  
                country_info_list.append(country_info[3].text)
                country_info_list.append(country_info[4].text)
                country_info_list.append(country_info[5].xpath("./span[1]/text()")[0])
                country_info_list.append(country_info[5].xpath("./span[2]/text()")[0])  
                country_info_list.append(country_info[6].text)
                country_info_list.append(time.text[3:-3])
                print(country_info_list)
            with open('weather.csv', 'a+', encoding='utf-8') as file:
                for m in range(0,len(country_info_list)):
                    file.write(country_info_list[m]+',')
                file.write('\n')
                
                
                
```
