```python
query = f'from(bucket: "{bucket}") |> range(start: -1h) |> filter(fn: (r) => r._measurement == "mem")'
tables = client.query_api().query(query, org=org)
tables
```

```python
results = []
for table in tables:
    for record in table.records:
        results.append((record.get_value(), record.get_field()))
print(results)
```
