```python
from lxml import etree
```

```python
selector = etree.HTML(res.text) #将源码转化为能被XPath匹配的格式
```

```python
data = selector.xpath("//div[@class='weilai']")[0]
```

```python
time = data.xpath("./div[2]/text()")[0]
```

```python
time
```

```python
time = time.split('：')[1]
```
